```python
import numpy as np
rng = np.random.default_rng()
arr = rng.standard_normal((5, 4))
# axis=0 son las columnas (hacia abajo a través de las filas)
print("Promedio por columna:", arr.mean(axis=0))
# axis=1 son las filas (a través de las columnas)
print("Suma por fila:", arr.sum(axis=1))
```
