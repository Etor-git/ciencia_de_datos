# Ejercicios Prácticos en Python


NumPy (Numerical Python), el paquete fundamental para la computación científica en Python. Su
principal característica es el ndarray, un objeto de arreglo multidimensional rápido y flexible que
permite realizar operaciones matemáticas en bloques enteros de datos con una sintaxis similar a la de los
escalares.
1. El objeto ndarray
 Homogeneidad: A diferencia de las listas de Python, todos los elementos de un ndarray deben
ser del mismo tipo (dtype).
 Eficiencia: NumPy almacena los datos en bloques continuos de memoria, lo que permite que
algoritmos escritos en C operen sin la sobrecarga de Python, resultando en velocidades de 10 a
100 veces superiores.
 Creación: Se pueden crear arreglos usando np.array() (desde listas), o funciones como
np.zeros(), np.ones() y np.arange().
2. Operaciones y Vectorización
 Aritmética: Las operaciones entre arreglos del mismo tamaño se aplican elemento a elemento.
 Broadcasting: Es el proceso por el cual NumPy maneja operaciones entre arreglos de diferentes
tamaños.
 Ufuncs (Funciones Universales): Son funciones que realizan operaciones rápidas elemento a
elemento, como np.sqrt(), np.exp() (unarias) o np.add() y np.maximum()
(binarias).
3. Indexación y Segmentación (Slicing)
 Vistas vs. Copias: Una distinción crítica es que los cortes (slices) de arreglos son vistas del
arreglo original. Esto significa que cualquier modificación en el corte se reflejará en los datos
originales para ahorrar memoria.
 Indexación Booleana: Permite filtrar datos usando comparaciones vectorizadas (ej. data[names
== "Bob"]).
 Fancy Indexing: Se refiere a la indexación mediante arreglos de enteros para seleccionar
subconjuntos en un orden específico. A diferencia del slicing, esto siempre crea una copia.
4. Métodos Matemáticos y Estadísticos
NumPy ofrece métodos integrados para procesar estadísticas de todo el arreglo o a lo largo de un eje
(axis):
 Reducciones: sum(), mean(), std() (desviación estándar).
 Acumulaciones: cumsum() (suma acumulada).
 Lógica Condicional: np.where(condición, valor_si_true,
valor_si_false) es una versión vectorizada de la expresión ternaria de Python.

# Ejercicio 2: Indexación y Vistas (Slicing)
Instrucciones: Demuestra la diferencia entre una vista y una copia.

```python
# 1. Crea un arreglo del 0 al 9
arr = np.arange(10)
# 2. Crea un slice de los elementos del índice 5 al 8 y cámbialos a 12
arr_slice = arr[5:8]
arr_slice = 12345
# 3. Observa el arreglo original
print("Arreglo original modificado por el slice:", arr)
# El índice 6 habrá cambiado
# 4. Ahora haz lo mismo pero usando .copy()
arr_copy = arr[5:8].copy()
arr_copy[:] = 0
print("Arreglo original tras modificar una copia:", arr) # No habrá cambiado
```

# Ejercicio 3: Indexación Booleana y np.where
Instrucciones: Dado un arreglo de datos, sustituye todos los valores negativos por 0 y los positivos por 1.

```python
rng = np.random.default_rng()
data = rng.standard_normal((4, 4))
print("Datos originales:\n", data)
# Usando np.where para lógica condicional
resultado = np.where(data > 0, 1, 0)
print("\nDatos binarizados (0 o 1):\n", resultado)
# Alternativa: Sustituir solo negativos por 0 en el arreglo original
data[data < 0] = 0
print("\nArreglo original filtrado (solo >= 0):\n", data)
```

# Ejercicio 4: Estadísticas por Ejes
Instrucciones: Crea una matriz de 5x4 y calcula el promedio de cada columna y la suma total de cada
fila.

```python
rng = np.random.default_rng()
arr = rng.standard_normal((5, 4))
# axis=0 son las columnas (hacia abajo a través de las filas)
print("Promedio por columna:", arr.mean(axis=0))
# axis=1 son las filas (a través de las columnas)
print("Suma por fila:", arr.sum(axis=1))
```

# Ejercicio 5: Álgebra Lineal Básica
Instrucciones: Realiza una multiplicación de matrices entre una matriz de 2x3 y su transpuesta.

```python
rng = np.random.default_rng()
X = rng.standard_normal((2, 3))
# Multiplicación de matriz por su transpuesta usando el operador @
resultado = X @ X.T
print("Resultado de X multiplicado por X transpuesta:\n", resultado)
# Alternativa usando np.dot
resultado_dot = np.dot(X, X.T)
```

## **Ejercicio 6: Creación con Tipos de Datos Específicos**

Instrucciones: Crea un arreglo de 3x3 lleno de ceros, pero asegúrate de que el tipo de dato sea entero de
32 bits (int32). Luego, cambia todos los elementos de la diagonal principal a 1.

```python
import numpy as np
# 1. Crear arreglo 3x3 de ceros int32
arr = np.zeros((3, 3), dtype=np.int32)
# 2. Cambiar la diagonal (usando indexing básico o np.eye como referencia)
arr[np.diag_indices(3)] = 1
print("Arreglo con diagonal de unos:\n", arr)
print("Dtype:", arr.dtype) # Debe ser int32
```

# Ejercicio 7: Filtrado Lógico Complejo

Instrucciones: Crea un arreglo con los números del 1 al 20. Utiliza indexación booleana para seleccionar solo los números que sean mayores que 10 y que sean pares.

```python
arr = np.arange(1, 21)
# Crear la máscara booleana combinando condiciones con &
mask = (arr > 10) & (arr % 2 == 0)
resultado = arr[mask]
print("Números > 10 y pares:", resultado)
```

# Ejercicio 8: Transformaciones y Transposición

Instrucciones: Crea un arreglo de 15 elementos (0 a 14) y cámbiale la forma (reshape) a una matriz de 3x5. Luego, obtén su transpuesta y explica qué sucede con la forma.

```python
#Crear un elemento
arr = np.arange(15).reshape((3, 5))
print("Forma original:", arr.shape) # (3, 5)
# Obtener la transpuesta usando el atributo .T
arr_t = arr.T
print("Forma de la transpuesta:", arr_t.shape) # Debe ser (5, 3)
```

# Ejercicio 9: Uso de Funciones Universales (ufuncs) Binarias

Instrucciones: Crea dos arreglos de 5 números aleatorios cada uno. Utiliza la función np.maximum para obtener un tercer arreglo que contenga el valor más alto de cada posición comparada.

```python
rng = np.random.default_rng(seed=123)
x = rng.standard_normal(5)
y = rng.standard_normal(5)
# np.maximum compara elemento a elemento
resultado = np.maximum(x, y)
print("X:", x)
print("Y:", y)
print("Máximos elemento a elemento:", resultado)
```
