# Proyecto 1: Resolver ecuaciones automáticamente

* Ecuaciones cuadráticas
* Factorización
* Raíces

¿Qué estamos viendo?

1. x = symbols('x')

Aquí definimos una variable simbólica.

Esto no es un número, es como en álgebra:

  x puede valer cualquier cosa


2. ecuacion = x**2 - 5*x + 6

Esto representa:

x^2 - 5x + 6 = 0

Esto es una ecuación cuadrática


3. solve(ecuacion, x)

Aquí lo convierte en formula a código

Resuelve:
x^2 - 5x + 6 = 0

Factorizando:
(x - 2)(x - 3) = 0

Entonces:

* x = 2
* x = 3

```python
from sympy import symbols, solve

x = symbols('x')
ecuacion = x**3 - 9*x + 7
sol = solve(ecuacion, x)

print("Soluciones:", sol)
```

# Proyecto 2: Análisis de datos simple

* Media
* Mediana
* Varianza

¿Qué estamos viendo?

Datos: 10, 20, 20, 30, 40

Media (promedio)

media = 10 + 20 + 20 + 30 + 40 / 5 = 24

Esto representa el “centro” de los datos

Mediana

Ordenados: 10, 20, 20, 30, 40
El valor del centro es: 20

Desviación estándar

Mide qué tanto se alejan los datos del promedio.

* Si es baja → datos similares
* Si es alta → datos dispersos

```python
import pandas as pd

data = [10, 20, 20, 30, 40]

df = pd.DataFrame(data, columns=['Valores'])

print("Media:", df.mean())
print("Mediana:", df.median())
print("Desviación estándar:", df.std())
```

# Proyecto 4: Predicción simple (regresión)

* Relación entre variables
* Introducción a ML

¿Qué estamos viendo?

Datos:
X: 1,2,3,4
y: 2,4,6,8

Relación:
y = 2x

modelo.fit(X, y)

El modelo aprende la relación:

Encuentra la recta:
y = mx + b

En este caso:

* m = 2
* b = 0

Predicción: x = 5 → y = 10

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

modelo = LinearRegression()
modelo.fit(X, y)

print("Predicción para 5:", modelo.predict([[5]]))
```
