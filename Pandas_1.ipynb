**Objetivo:**
Practicar la creación de una Serie asignando etiquetas específicas a los datos y seleccionando elementos mediante esas etiquetas.

**Instrucciones:**
Crea una Serie de pandas con los valores
 y asígnales los índices ['Venta A', 'Venta B', 'Venta C', 'Venta D'].

**Objetivo:**

**Instrucciones:**
Crea una Serie llamada tienda1 a partir del diccionario: {'manzanas': 10, 'peras': 20, 'plátanos': 15}.

```python
import pandas as pd
```

```python
#Crea series desde diccionarios
tienda1 = pd.Series({'manzanas': 10, 'peras': 20, 'plátanos': 15})
tienda2 = pd.Series({'peras': 5, 'plátanos': 10, 'uvas': 30})
print("Tienda 1:\n", tienda1)
print("\nTienda 2:\n", tienda2)
```

**Instrucciones:**
Dada una Serie de temperaturas: pd.Series(
, index=['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb']).

Filtra la Serie para mostrar solo los días donde la temperatura fue superior a 23 grados.

Multiplica todos los valores de la Serie por 1.8 y suma 32 (conversión a Fahrenheit) y muestra el resultado.

```python
import pandas as pd
```

```python
temp_c = pd.Series([20,22,25,23,27,21], index=['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb'])
print("Temperaturas en Celsius:\n", temp_c)
```

```python
print("Días calurosos (>23°C):\n", temp_c[temp_c <= 22])
```

**Ejercicio 4: **

Manejo de Datos Faltantes (Limpieza)
Objetivo: Utilizar las herramientas de pandas para identificar y tratar valores nulos (NaN o null), una tarea crítica en la preparación de datos.

**Instrucciones:**
Crea una Serie que contenga los valores [1.5, None, 3.0, None, 5.5].
Utiliza el método isna() para mostrar una máscara booleana de los valores faltantes.

Crea una copia de la Serie donde los valores faltantes sean reemplazados por el promedio de los valores existentes usando fillna().

Crea otra copia donde simplemente se eliminen las entradas nulas usando dropna()
.

```python
import numpy as np
```

```python
datos = pd.Series([1.5, np.nan, 3.0, np.nan, 5.5])
print("Serie original:\n", datos)
```

```python
#Detectar nulos
print("¿Es nulo?:\n", datos.isna())
```

```python
promedio = datos.mean()
datos_rellenos = datos.fillna(promedio)
print("\nSerie tras fillna (con promedio):\n", datos_rellenos)
```

```python
datos_limpios =datos.dropna()
print("\nSerie tras dropna:\n", datos_limpios)
```
