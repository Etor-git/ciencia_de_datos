```python
import numpy as np
# 1. Crear el generador aleatorio
rng = np.random.default_rng(seed=42)
# 2. Crear un arreglo 3x4
data = rng.standard_normal((3, 4))
print("Arreglo original:\n", data)
# 3. Multiplicar todos los elementos por 10 y convertir a enteros (int32)
data_int = (data * 10).astype(np.int32)
print("\nArreglo transformado (int32):\n", data_int)
```
