**Ejercicio 2**
Agrupación y Estadística Descriptivas
**Objetivo:** Aplicar la metodología split-apply-combine para resumir datos numéricos por categorías.
Calcula el promedio del "Engagement Score", "Satisfaction Score", y "Work-Life Balance Score" para cada unidad.
Ordena los resultados de mayor a menor basándote en el "Engagement Score" para identificar la unidad con empleados más comprometidos.

```python
metricas_unidad = df.groupby("BusinessUnit")[["Engagement Score", "Satisfaction Score", "Work-Life Balance Score"]].mean()
```

```python
ranking_compromiso = metricas_unidad.sort_values(by="Engagement Score", ascending=False)
#print(ranking_compromiso)
ranking_compromiso
```
