```python
import pandas as pd
from numpy import dtype
```

Una serie en la libreria pandas es como una columna en una tabla

```python
a = [1,7,2]
print("Este es el arreglo de a \n", a)
```

```python
#Convertir la variable a en una serie pandas
a = pd.Series(a)
print("Este es el arreglo de a \n", a)
```

```python
print("Este es el primer elemento de la serie a es:\n", a[0])
print("Este es el primer elemento de la serie a es:\n", a[1])
print("Este es el primer elemento de la serie a es:\n", a[2])
```

Agregamos un indice para la serie

```python
a = pd.Series(a, index = ["x","y","z"])
print(a)
```

```python
a["x"] = 1
a["y"]= 7
a["z"]= 2
print(a["x"])
print(a["y"])
print(a["z"])
```

Confirmamos que tipo de dato se trata

```python
dtype(a)
```

Como creamos un Dataframe apartir de un diccionario

```python
#Creo un diccionario con las key dia1 a dia3
calorias = {"dia1": 420, "dia2": 380, "dia3": 390}

#Usando la funcion Series de pandas para convertirlo a serie
calorias = pd.Series(calorias)

#Imprimimos la serie calorias
print(calorias)
```

Se define técnicamente como un objeto bidimensional con etiquetas en sus ejes (filas y columnas)

```python
#Creamos un Dataframe apartir de dos
data = {
    "calorias": [420, 380, 390],
    "duracion": [50, 40, 45]
}
print(data)
```

```python
#Creo un Dataframe apartir de la variable data
df = pd.DataFrame(data)
print(df)
```
